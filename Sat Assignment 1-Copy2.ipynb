```python
b['Goal_diff_count']=0
b['GoalsFor'].dtypes
pd.to_numeric(b['GoalsFor'])
pd.to_numeric(b['GoalsAgainst'])
b['GoalsFor']=b['GoalsFor'].astype(int)
b['GoalsAgainst']=b['GoalsAgainst'].astype(int)
b['Goal_diff_count']=b['GoalsFor']-b['GoalsAgainst']
b
```

```python
d=b['Goal_diff_count'].max()
b[b['Goal_diff_count']==d]
```

```python
#Create a new column with name “Winning Percent” and append it to the data set (5 points)
#Percentage of Winning = (GamesWon / GamesPlayed)*100

#If there are any numerical error, replace it with 0%

#Print the top 5 teams which has the highest Winning percentage

b['Winning Percent']=0
pd.to_numeric(b['GamesWon'])
pd.to_numeric(b['GamesPlayed'])
b['GamesWon']=b['GamesWon'].astype(int)
b['GamesPlayed']=b['GamesPlayed'].astype(int)
b['Winning Percent']=(b['GamesWon']/b['GamesPlayed'])*100
b=b.fillna(value=0)
b
```
